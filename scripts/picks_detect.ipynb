```python
import pandas as pd
import numpy as np
import os
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../database/recordings_processed/principiantes/20240507_113717_pablo/Xsens DOT_1_D422CD006775_20240507_113115.csv')
#df = pd.read_csv('../database/recordings_processed/avanzados/20240507_151249_andreaa_1/Xsens DOT_1_D422CD006775_20240507_151203.csv')
f = 60#Hz
df['ElapsedTime'] = (df['PacketCounter']-1)*(1/f)

#guardar csv
df.to_csv('../database/Xsens DOT_1_D422CD006775_20240507_113115_con_tiempo.csv', index=False)
df.head()
```

```python
import plotly.express as px
px.scatter(df,x='ElapsedTime',y='Acc_X')
```

```python
maximos = [-100]
indices = [0]
picos = []
indices_picos = []
```

```python
for i in range(1,len(df)):
    if(df['Acc_X'][i]>df['Acc_X'][i-1]):
        maximos.append(df['Acc_X'][i])
        indices.append(i)
    else:
        if(df['Acc_X'][i]<(maximos[-1]-2) and df['ElapsedTime'][i]-df['ElapsedTime'][indices[-1]]>0.1):
            picos.append(maximos[-1])
            indices_picos.append(indices[-1])
            maximos = [-100]
            indices = [i]
```

```python
print(picos)
len(picos)
```

```python
import plotly.express as px
import plotly.graph_objects as go
#graficar picos
fig = px.scatter(df,x='ElapsedTime',y='Acc_X')
fig.add_trace(go.Scatter(x=df['ElapsedTime'][indices_picos], y=picos, mode='markers', marker=dict(color='red', size=10)))
fig.show()

```

```python
#select 10 highest peaks
picos = np.array(picos)
indices_picos = np.array(indices_picos)
indices_picos = indices_picos[picos.argsort()[-10:]]
picos = picos[picos.argsort()[-10:]]
print(picos)
```

```python
#plot peaks
fig = px.scatter(df,x='ElapsedTime',y='Acc_X')
fig.add_trace(go.Scatter(x=df['ElapsedTime'][indices_picos], y=picos, mode='markers', marker=dict(color='red', size=10)))
fig.show()
```
